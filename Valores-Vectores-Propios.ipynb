# Valores y Vectores Propios

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

%matplotlib inline
```

```python
def plot_vect(x, b, xlim, ylim):
    '''
    funcion para graficar dos vectores,
    x - el vector original
    b - el vector transformado
    xlim - el limite para x
    ylim - el limite para y
    '''
    plt.figure(figsize = (10, 6))
    plt.quiver(0,0,x[0],x[1],\
        color='k',angles='xy',\
        scale_units='xy',scale=1,\
        label='Vector Original')
    plt.quiver(0,0,b[0],b[1],\
        color='g',angles='xy',\
        scale_units='xy',scale=1,\
        label ='Vector Transformado')
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```

### Grafico de dos vectores

```python
A = np.array([[2, 0],[0, 1]])

x = np.array([[1],[1]])
b = np.dot(A, x)

plot_vect(x,b,(0,3),(0,2))
```

### Grafico del vector transformado

```python
x = np.array([[1], [0]])
b = np.dot(A, x)

plot_vect(x,b,(0,3),(-0.5,0.5))
```

## El metodo de potencias

```python
def normalize(x):
    fac = abs(x).max()
    x_n = x / x.max()
    return fac, x_n
```

```python
x = np.array([1, 1])
a = np.array([[0, 2], 
              [2, 3]])

for i in range(8):
    x = np.dot(a, x)
    lambda_1, x = normalize(x)
    
print('Valor propio:', lambda_1)
print('Vector propio:', x)
```

### El metodo de potenicas inverso

```python
from numpy.linalg import inv
```

```python
a_inv = inv(a)

for i in range(8):
    x = np.dot(a_inv, x)
    lambda_1, x = normalize(x)
    
print('Valor propio:', lambda_1)
print('Vector propio:', x)
```

### El metodo QR

```python
from numpy.linalg import qr
```

```python
a = np.array([[0, 2], 
              [2, 3]])

q, r = qr(a)

print('Q:', q)
print('R:', r)

b = np.dot(q, r)
print('QR:', b)
```

```python
a = np.array([[0, 2], 
              [2, 3]])
p = [1, 5, 10, 20]
for i in range(20):
    q, r = qr(a)
    a = np.dot(r, q)
    if i+1 in p:
        print(f'Iteracion {i+1}:')
        print(a)
```

### Valores y vectores propios

```python
from numpy.linalg import eig
```

```python
a = np.array([[0, 2], 
              [2, 3]])
w,v = eig(a)
print('E-valor:', w)
print('E-vector', v)
```

```python
a = np.array([[2, 2, 4], 
              [1, 3, 5],
              [2, 3, 4]])
w,v = eig(a)
print('E-value:', w)
print('E-vector', v)
```
